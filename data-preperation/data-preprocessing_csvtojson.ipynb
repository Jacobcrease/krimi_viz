```python
import pandas as pd
import os
```

```python
df1 = pd.read_csv("./data_csv/FaelleZeitreihe.csv", sep=";")
df1.head()
```

```python
df2 = pd.read_csv("./data_csv/ZR-F-01-T01-Faelle_csv.csv",
                  sep=";",
                  #decimal=",",
                  encoding="latin_1",
                  header=1)

df2.replace(',','', regex=True, inplace=True)
df2.head()
```

```python
for col in list(df2.columns)[2:]:
    df2[col] = pd.to_numeric(df2[col],  errors='coerce')
```

```python
df2_store_path = "data_json/ZR-F-01-T01-Faelle_json.json"
df2_full_path = os.path.join(os.getcwd(), df2_store_path)

f = open(df2_full_path, "w")
f.write("")
f.close()

df2.to_json(df2_full_path, orient="records", force_ascii=False)
```

```python
# Liste mit allen erfassten Straftaten, sortiert nach Fallzahl insgesamt
df2[["Straftat", "erfasste Faelle"]].groupby("Straftat").sum("erfasste Faelle").sort_values("erfasste Faelle", ascending=False).head(30)
```

```python
# Liste aller Straftaten, welche "straftat_str" inbehalten, sortiert nach Fallzahl insgesamt

"""
Straftatenliste:
"Gewaltkriminalität"
"Rauschgiftkriminalität darunter:"
"Sachbeschädigung §§ 303-305a StGB darunter:" + "Sachbeschädigung §§ 303-305a StGB"
contains "Betrug §§ 263" + "sonstiger Betrug darunter:" + contains "Sonstiger Betrug"
"Diebstahl insgesamt" + "Diebstahl insgesamt und zwar:"
"Sonstige Straftatbestände (StGB)"
"""

straftat_str = "Diebstahl"

df2[["Straftat", "erfasste Faelle"]][df2["Straftat"].str.contains(straftat_str)].groupby("Straftat").sum("erfasste Faelle").sort_values("erfasste Faelle", ascending=False)
```

```python
# Auflistung jährlicher Fallzahlen für straftat_str
straftat_str = "Straßenkriminalität"
df2[["Straftat", "Jahr", "erfasste Faelle"]][df2["Straftat"] == straftat_str]
```

```python
# Auflistung jährlicher Fallzahlen für straftat_str
straftat_str = "Betrug"
df2[["Straftat", "Jahr", "erfasste Faelle"]][df2["Straftat"].str.contains(straftat_str)][df2["Jahr"]>2013].groupby("Straftat").sum("erfasste Faelle").sort_values("erfasste Faelle", ascending=False)
```

```python
# Tabelle generieren mit Spalten [Straftat, Jahr, Fallzahlen]

"""
Straftatenliste:
"Gewaltkriminalität"
"Rauschgiftkriminalität darunter:"
"Sachbeschädigung §§ 303-305a StGB darunter:" + "Sachbeschädigung §§ 303-305a StGB"
contains "Betrug §§ 263" + "sonstiger Betrug darunter:" + contains "Sonstiger Betrug"
"Diebstahl insgesamt" + "Diebstahl insgesamt und zwar:"
"Sonstige Straftatbestände (StGB)"
"""

df3_1 = pd.DataFrame()
df3_1 = df2[["Straftat", "Jahr", "erfasste Faelle"]][df2["Straftat"] == "Gewaltkriminalität"][df2["Jahr"] >= 1993]

df3_2 = pd.DataFrame()
df3_2 = df2[["Straftat", "Jahr", "erfasste Faelle"]][df2["Straftat"] == "Rauschgiftkriminalität darunter:"][df2["Jahr"] >= 1993]

df3_2["Straftat"] = "Rauschgiftkriminalität"

df3_3 = pd.DataFrame()
df3_3 = df2[["Straftat",
             "Jahr",
             "erfasste Faelle"]][df2["Straftat"].isin(["Sachbeschädigung §§ 303-305a StGB darunter:",
                                                       "Sachbeschädigung §§ 303-305a StGB"])][df2["Jahr"] >= 1993]

df3_3["Straftat"] = "Sachbeschädigung"

df3_4_a = pd.DataFrame()
df3_4_a = df2[["Straftat", "Jahr", "erfasste Faelle"]][df2["Straftat"].str.contains("Betrug §§ 263")][df2["Jahr"] >= 1993]

df3_4_a["Straftat"] = "Betrug"

df3_4_b = pd.DataFrame()
df3_4_b = df2[["Straftat", "Jahr", "erfasste Faelle"]][df2["Straftat"] == "sonstiger Betrug darunter:"][df2["Jahr"] >= 1993]

df3_4_b["Straftat"] = "Betrug"

df3_4_c = pd.DataFrame()
df3_4_c = df2[["Straftat", "Jahr", "erfasste Faelle"]][df2["Straftat"].str.contains("Sonstiger Betrug")][df2["Jahr"] >= 1993]

df3_4_c["Straftat"] = "Betrug"

df3_4 = pd.concat([df3_4_a, df3_4_b, df3_4_c]).groupby(["Straftat", "Jahr"]).sum().reset_index()

df3_5 = pd.DataFrame()
df3_5 = df2[["Straftat",
             "Jahr",
             "erfasste Faelle"]][df2["Straftat"].isin(["Diebstahl insgesamt",
                                                       "Diebstahl insgesamt und zwar:"])][df2["Jahr"] >= 1993]

df3_5["Straftat"] = "Diebstahl"

df3_6 = pd.DataFrame()
df3_6 = df2[["Straftat", "Jahr", "erfasste Faelle"]][df2["Straftat"] == "Sonstige Straftatbestände (StGB)"][df2["Jahr"] >= 1993]

df3_6["Straftat"] = "Sonstige"

df3 = pd.concat([df3_1, df3_2, df3_3, df3_4, df3_5, df3_6]).groupby(["Straftat", "Jahr"]).sum().reset_index()
df3
```

```python
df3_store_path = "data_json/ZR-F-01-T01-Faelle_filtered_json.json"
df3_full_path = os.path.join(os.getcwd(), df3_store_path)

f = open(df2_full_path, "w")
f.write("")
f.close()

df3.to_json(df3_full_path, orient="records", force_ascii=False)
```
